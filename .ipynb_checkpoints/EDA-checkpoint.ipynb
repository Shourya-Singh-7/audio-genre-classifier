```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = "data/genres_original/blues/blues.00000.wav"

y, sr = librosa.load(path, sr=None, mono=True)  # sr=None keeps original sr

print("Shape:", y.shape)
print("Sample rate:", sr)
print("Duration (sec):", len(y) / sr)
```

```python
plt.figure(figsize=(15, 5))
librosa.display.waveshow(y, sr=sr)
plt.title("Waveform")
plt.show()
```

```python
S = np.abs(librosa.stft(y))
S_db = librosa.amplitude_to_db(S, ref=np.max)

plt.figure(figsize=(14, 4))
librosa.display.specshow(S_db, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format="%+2.0f dB")
plt.title("Spectrogram (dB, Log Frequency)")
plt.show()
```

```python
import os
durations = []

for root, dirs, files in os.walk("data/genres_original"):
    for f in files:
        if f.endswith(".wav"):
            y, sr = librosa.load(os.path.join(root, f), sr=None)
            durations.append(len(y)/sr)

plt.hist(durations, bins=20)
plt.title("Audio Duration Distribution")
plt.show()
```
